```python
def plot_ts_and_points(ts, start_point, step):
    new_series = [None for i in range(len(ts))]
    for i in range(len(ts)):
        pos = start_point + step * i
        if pos >= len(ts):
            break
        new_series[pos] = ts[pos]
    new_series = pd.Series(new_series)
    
    with plt.style.context('bmh'):
        plt.figure(figsize=(16, 8))
        ts_ax = plt.axes()
        ts.plot(ax=ts_ax, color='blue')
        new_series.plot(ax=ts_ax, style='ro')
```

# Тест Дики-Фуллера 
Проверяем гипотезу не стационарности против стационарности

```python
from statsmodels.tsa.stattools import adfuller
```

```python
################
 # Dickey-Fuller
##################
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
test_stationarity(series_auto)
```

```python
test_stationarity(series_dj)
```

по тесту оба графика стационарные , т.к. p-value <.05
посмотрим на коррелограммы , чтобы подтвердить тест

# Коррелограммы

```python
def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05) # полная автокорреляция 
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05) # частичная автокорреляция

        plt.tight_layout()
    return
```

```python
tsplot(series_auto)
```

# Work with Non - Stationary Series

### Хольт-Винтерс

Важно - метод применим только в случае сезонности. Если ее нет - метод хольта-винтерса не даст хороших результатов

```python
# используем в качестве тестового и валидационного ряда период в 6 месяцев
train_nonst, test_nonst, val_nonst = series_auto_notst[:-12], series_auto_notst[-12:-6], series_auto_notst[-6:]
```

```python
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error
```

```python
def initial_trend(series, season_len):
    return sum([float(series[i + season_len] - series[i]) / season_len]) / season_len
```

```python
def initial_seasonal_components(series, slen):
    seasonals = {}
    season_averages = []
    n_seasons = int(len(series)/slen)
    # compute season averages
    for j in range(n_seasons):
        season_averages.append(sum(series[slen*j:slen*j+slen])/float(slen))
    # compute initial values
    for i in range(slen):
        sum_of_vals_over_avg = 0.0
        for j in range(n_seasons):
            sum_of_vals_over_avg += series[slen*j+i]-season_averages[j]
        seasonals[i] = sum_of_vals_over_avg/n_seasons
    return seasonals
```

```python
def triple_exponential_smoothing(series, slen, alpha, beta, gamma, n_preds):
    result = []
    seasonals = initial_seasonal_components(series, slen)
    for i in range(len(series)+n_preds):
        if i == 0: # initial values
            smooth = series[0]
            trend = initial_trend(series, slen)
            result.append(series[0])
            continue
        if i >= len(series): # we are forecasting
            m = i - len(series) + 1
            result.append((smooth + m*trend) + seasonals[i%slen])
        else:
            val = series[i]
            last_smooth, smooth = smooth, alpha*(val-seasonals[i%slen]) + (1-alpha)*(smooth+trend)
            trend = beta * (smooth-last_smooth) + (1-beta)*trend
            seasonals[i%slen] = gamma*(val-smooth) + (1-gamma)*seasonals[i%slen]
            result.append(smooth+trend+seasonals[i%slen])
    return result
```

Подберем оптимальные коэффициенты alpha, beta, gamma

```python
def mse_nst(X):
    alpha, beta, gamma = X
    result = triple_exponential_smoothing(train_nonst, 12, alpha, beta, gamma, len(test_nonst))
    predictions = result[-len(test_nonst):]
    error = mean_squared_error(predictions, test_nonst)
    return error
```

```python
opt = minimize(mse_nst, x0=[0,0,0], method="L-BFGS-B", bounds = ((0.05, .9), (0.05, .9), (0.05, .9)))
```

```python
alpha_opt, beta_opt, gamma_opt = opt.x
print(opt)
```

```python
opt['x']
```

```python
def plot_tema(alpha, beta, gamma, ser=train_nonst, ser_to_plot=series_auto_notst[:-6], n_preds=len(val_nonst)):
    tema = triple_exponential_smoothing(ser, 12, alpha, beta, gamma, n_preds)
    with plt.style.context('bmh'):
        plt.figure(figsize=(14, 8))
        plt.plot(ser_to_plot, color='blue',label='original')
        plt.plot(tema, color='red', linewidth='1', label='TEMA')
        plt.title("alpha={}, beta={}, gamma={}".format(alpha, beta, gamma))
        plt.legend()
```

```python
predictions_tema = triple_exponential_smoothing(series_auto_notst[:-6], 12, alpha_opt, beta_opt,\
                                                gamma_opt, len(val_nonst))
```

# Work with Stationary Series

```python
train_st, test_st, val_st = series_auto_st[:-12], series_auto_st[-12:-6], series_auto_st[-6:]
```

## Скользящее среднее

```python
def predict_ma(series, window, n_pred):
    new_series = series.copy()
    new_series = pd.Series(new_series)
    result = new_series.rolling(window).mean()
    for _ in range(n_pred):
        result = result.append(pd.Series([new_series[-window:].mean()]), ignore_index=True)
    return result
```

```python
def mse_ma(window,train,test):
    result = predict_ma(train, window, len(test))
    predictions = result[-len(test):]
    error = mean_squared_error(predictions, test)
    return error
```

```python
predictions_ma = predict_ma(series_auto_st[:-6], 9, len(val_st))
```

```python
ma_error = mse_ma(9, series_auto_st[:-6], val_st)
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(14, 8))
    plt.plot(series_auto_st, color='blue',label='original')
    plt.plot(predictions_ma, color='red', linewidth='4', label='MA')
    plt.legend()
```

Предсказания ma на валидационном ряду - горизонтальная линия.

Чем больше размер "окна" скользящего среднего $n$ тем более сглаженным получается результат. По нескольким скользящим средним можно увидеть периоды резкого роста и резкого спада.

```python
print("MSE = {}".format(ma_error))
```

# Взвешенная скользящая средняя - (WMA)
weights - степень двойки

```python
def predict_wma(series, window, n_pred):
    weights = np.array([i**2 for i in range(1,window+1)])
    new_series = series.copy()
    new_series = pd.Series(new_series)
    result = new_series.rolling(window).apply(lambda s: (s * weights).sum() / weights.sum(), raw=True)
    for _ in range(n_pred):
        result = result.append(pd.Series([(np.array(new_series[-window:])*weights/weights.sum()).sum()]),ignore_index = True)
    return result
```

```python
def mse_wma(window, train, test):
    result = predict_wma(train, window, len(test))
    predictions = result[-len(test):]
    error = mean_squared_error(predictions, test)
    return error
```

```python
predictions_wma = predict_wma(series_auto_st[:-6], 9, len(val_st))
```

```python
wma_error = mse_wma(9, series_auto_st[:-6], val_st)
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(14, 8))
    plt.plot(series_auto_st, color='blue',label='original')
    plt.plot(predictions_wma, color='red', linewidth='4', label='WMA')
    plt.legend()
```

```python
print("MSE = {}".format(wma_error))
```

# Экспоненциальное сглаживание - (EMA)

```python
def predict_ema(series, alpha, n_pred):
    result = [series[0]]
    for n in range(1, len(series)):
        result.append(alpha * series[n-1] + (1 - alpha) * result[n-1])
    
    predictions = [alpha * series[len(series) - 1] + (1 - alpha) * result[-1]]

    for _ in range(n_pred):
        #res = alpha * predictions[-1] + (1 - alpha) * result[-1]
        result.append(alpha * predictions[-1] + (1 - alpha) * result[-1])
        predictions.append(alpha * predictions[-1] + (1 - alpha) * result[-1])
    
    return pd.Series(result)
```

```python
def mse_ema(X, train, test):
    alpha = X
    result = predict_ema(train, alpha, len(test))
    predictions = result[-len(test_st):]
    error = mean_squared_error(predictions, test)
    return error
```

```python
predictions_ema = predict_ema(series_auto_st[:-6], 0.2, len(val_st))
```

```python
ema_error = mse_ema(0.2, series_auto_st[:-6], val_st)
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(14, 8))
    plt.plot(series_auto_st, color='blue',label='original')
    plt.plot(predictions_ema, color='red', linewidth='4', label='EMA')
    plt.legend()
```

```python
print("MSE = {}".format(ema_error))
```

Для стационарного ряда MA, WMA и EMA показали приблизительно один и тот же результат.
