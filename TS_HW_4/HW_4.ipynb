```python
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.simplefilter('ignore')
p=print
```

```python
all_text = ''
for i in range(1, 5):
    with open(f'tom_{i}.txt', 'r') as f:
        text = f.read().lower()
    text = re.sub('\s+', ' ', text)
    text = re.sub('[^а-я ]','', text)
    text = re.sub('\s+', ' ', text)
    all_text += ' ' + text
```

```python
RUSSIAN = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
POS_RU = {position: leter for leter, position in enumerate(RUSSIAN)}
probabilities_ru = np.zeros((len(RUSSIAN), len(RUSSIAN)))
```

```python
from collections import Counter
```

```python
for pair, count in Counter(zip(all_text[:-1], all_text[1:])).items():
    if ' ' not in pair:
        probabilities_ru[POS_RU[pair[0]], POS_RU[pair[1]]] = count
```

```python
probabilities = (probabilities_ru / probabilities_ru.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
```

```python
# here is matrix
plt.figure(figsize=(14,8))
sns.heatmap(df,annot=True,fmt='.2f')
plt.show()
```

```python
#Посчитайте вероятность встретить строку из 4-х символов, которая начинается п, а кончается на р
```

```python
start = POS_RU['п']
end = POS_RU['р']
prior = 1.0 / len(RUSSIAN)
prob = 0
for leter1 in RUSSIAN:
    for leter2 in RUSSIAN:
        prob += prior * (probabilities[start, POS_RU[leter1]] *probabilities[POS_RU[leter1], POS_RU[leter2]] *\
                         probabilities[POS_RU[leter2], end])
p('Вероятность встретить строку из 4-х символов, которая начинается п, а кончается на р = ',prob)        
```

```python
# Используя матрицу переходных вероятностей сгенерируйте слово, начинающееся с буквы “Б” и имеющее длину 8-мь символов
```

```python
s = ['б']
length = 8
for i in range(length):
    index = POS_RU[s[i]]
    next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
    s.append(next_char)

print('Слово - {}'.format(''.join(s)))
```

```python
data = pd.read_csv('numbers-on-unemployment-benefits.csv')
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(16, 8))
    layout = (3, 2)
    ts_ax = plt.subplot()
    data['Numbers Of Unemployment'].plot(ax=ts_ax, color='blue')
    ts_ax.set_title("Numbers-On-Unemployment-Benefits")
    plt.tight_layout()
```

```python
from hmmlearn.hmm import GaussianHMM
```

```python
values = data['Numbers Of Unemployment'].values.reshape(-1,1)
```

```python
score = {}
for n in range(1, 25):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    score[n] = model.score(values)
```

```python
p('Оценка модели \n {} '.format(score))
```

```python
fig = plt.figure(figsize=(12,6))
ax = plt.plot(score.keys(), score.values())
ticks = plt.xticks(range(1, 24))
```

```python
# будем брать 6 компонент
```

```python
components = 6
```

```python
model = GaussianHMM(n_components=6,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3);
```

```python
plt.figure(figsize=(16,12))
plt.plot(labels)
plt.show()
```

```python
len(model.transmat_)
```

```python
cur_value = labels[-20] # 20 - кол-во предсказаний 
predictions = []

for _ in range(20):
    cur_value = np.random.choice(a = components, p=model.transmat_[cur_value])    
    predictions.append(model.means_[cur_value])
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(16, 8))
    layout = (3, 2)
    ts_ax = plt.subplot()
    plt.plot(values[-20:])
    plt.plot(predictions,linewidth=3)
    ts_ax.set_title("Numbers-On-Unemployment-Benefits")
    plt.tight_layout()
```
